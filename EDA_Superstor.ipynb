# Data type conversion

```python
# Convert 'Date' column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format= '%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format= '%d-%m-%Y')

#Check convversion
print("\nData columns converted\n")
print(f"Order Date : {df['Order Date'].dtypes}")
print(f"Ship Date : {df['Ship Date'].dtypes}")
```

# Univariate Analysis

```python
# Univariate Analysis is the simplest form of analyzing data.As the name suggests- "unit"
# means "one", and "variate" means "variable". In univariate analysis, we analyze a single variable.

print("\n === CATAGORIAL VARIABLE ===\n")
catagorical_cols = ['Customer Segment', 'Category', 'Region', 'Payment Mode', 'Delivery Status', 'Auto Reorder']

for col in catagorical_cols:

    print(f"\n ---{col}---")
    print(df[col].value_counts())

    #visualization
    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(kind='bar', color = 'skyblue')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=35)
    plt.tight_layout()
    plt.show()
```

# Numerical Variable Analysis

```python
print("\n=== NUMERICAL VARIABLE ===\n")

numerical_cols = ['Quantity', 'Unit Price', 'Discount (%)', 'Sales Amount', 'Cost Price','Profit', 'Stock Left', 'Reorder Quantity']

for col in numerical_cols:

    print(f'\n--- {col}---\n')
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Std Dev: {df[col].std():.2f}")
    print(f"Min: {df[col].min():.2f}")
    print(f"Max: {df[col].max():.2f}")

    #visualization
    plt.figure(figsize=(8,4))
    
    plt.subplot(1,2,1)
    df[col].hist(bins=20, color='lightgreen', edgecolor='black')
    plt.title(f'Distribution of {col}')

    plt.subplot(1,2,2)
    df.boxplot(column = col)
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

# BIVARIATE ANALYSIS

```python
# Bivariate Analysis is the statistical analysis of two variables (independent and dependent) at the same time.
# the core purpose is to invesigate the relationship between the two variables.

print("\n ==== SALES & PROFIT BY CATEGORY ==== \n")

category_performance = df.groupby('Category').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

category_performance['Profit Margin'] = (category_performance['Profit'] / category_performance['Sales Amount']) * 100
print(category_performance,"\n\n")

# Visualization
fig,axes = plt.subplots(2,2, figsize=(20,15))

# Sales Amount by Category
category_performance['Sales Amount'].plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Total Sales Amount by Category')
axes[0,0].tick_params(rotation=35)

# Profit by Category
category_performance['Profit'].plot(kind='bar', ax=axes[0,1], color='green')
axes[0,1].set_title('Total Profit by Category')
axes[0,1].tick_params(rotation=35)

# Profit Margine
category_performance['Profit Margin'].plot(kind='bar', ax=axes[1,0], color='salmon')
axes[1,0].set_title('Profit Margin by Category')
axes[1,0].tick_params(rotation=35)

# Order Count
category_performance['Order Count'].plot(kind='bar', ax=axes[1,1], color='red')
axes[1,1].set_title('Order Count by Category')
axes[1,1].tick_params(rotation=35)


plt.tight_layout()
plt.show()
```

# Regional Performance

```python
print("\n ==== REGIONAL PERFORMANCE ==== \n")

regional_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

regional_performance['Profit Margin'] = (regional_performance['Profit'] / regional_performance['Sales Amount']) * 100
print(regional_performance,"\n\n")

# Visualization
plt.figure(figsize=(18,10))

plt.subplot(2,2,1)
regional_performance['Sales Amount'].plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Total Sales Amount by Region')

plt.subplot(2,2,2)
regional_performance['Profit'].plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Total Profit by Region')

plt.subplot(2,2,3)
sns.barplot(x = regional_performance.index, y = 'Order Count', data=regional_performance)
plt.title('Order Count by Region')
plt.xticks(rotation=35)

plt.subplot(2,2,4)
sns.barplot(x = regional_performance.index, y= 'Profit Margin', data=regional_performance)
plt.title('Profit Margin by Region (%)')
plt.xticks(rotation = 35)

plt.tight_layout()
plt.show()
```

# TIME SERIES ANALYSIS

```python
print("\n ==== TIME SERIES ANALYSIS ==== \n")

# Extracting month and year from 'Order Date'
df['Order Month'] = df['Order Date'].dt.to_period('M')
df['Order Year'] = df['Order Date'].dt.year

# Monthly Sales and Profit
monthly_trends = df.groupby('Order Month').agg({
    'Sales Amount' : 'sum',
    'Profit' : 'sum',
    'Order ID' : 'count'
}).rename(columns={'Order ID': 'Order Count'})

print("Monthly Trends:")
print(monthly_trends.tail(10),"\n\n")

# Visualization
plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
monthly_trends['Sales Amount'].plot(kind="line", marker='o', color='blue')
plt.title('Monthly Sales Amount Over Time')
plt.ylabel('Sales Amount')
plt.grid(True)

plt.subplot(3,1,2)
monthly_trends['Profit'].plot(marker='o', color='red')
plt.title('Monthly Profit Over Time')
plt.ylabel('Profit')
plt.grid(True)

plt.subplot(3,1,3)
monthly_trends['Order Count'].plot(marker='o', color='green')
plt.title('Monthly Order Count Over Time')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Advance Analysis

```python
print("\n=== CORRELATION ANALYSIS ===\n")

#Select numerical columns for correlation analysis
numerical_df = df[['Quantity', 'Unit Price', 'Discount (%)', 'Sales Amount', 'Cost Price','Profit', 'Stock Left', 'Reorder Quantity']]

# Calculate correlation matrix
correlation_matrix = numerical_df.corr()
print("Corrrelation Matrix:\n", correlation_matrix,"\n\n")

# Heatmap Visualization
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot= True, cmap= 'coolwarm', center = 0, square= True, linewidths= 0.5)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

# Customer Segment Analysis

```python
print("\n ==== COSTOMER SEGMENT ANALYSIS ==== \n")

segment_analysis = df.groupby('Customer Segment').agg({
    'Sales Amount': ['sum','mean'],
    'Profit': ['sum','mean'],
    'Order ID': 'count',
    'Discount (%)': 'mean'
}).round(2)

segment_analysis.columns = ['Total Sales', 'Avg Sales', 'Total Profit', 'Avg Profit', 'Order Count', 'Avg Discount (%)']
print(segment_analysis,"\n\n")

# Visualization
fig, axes = plt.subplots(2,2, figsize=(15,10))

segment_analysis['Total Sales'].plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Total Sales by Customer Segment')
axes[0,0].tick_params(axis='x', rotation=35)

segment_analysis['Total Profit'].plot(kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('Total Profit by Customer Segment')
axes[0,1].tick_params(axis='x', rotation=35)

segment_analysis['Avg Profit'].plot(kind='bar', ax=axes[1,0], color='salmon')
axes[1,0].set_title('Order Count by Customer Segment')
axes[1,0].tick_params(axis='x', rotation=35)

segment_analysis['Avg Discount (%)'].plot(kind='bar', ax=axes[1,1], color='pink')
axes[1,1].set_title('Average Discount by Customer Segment')
axes[1,1].tick_params(axis='x', rotation=35)

plt.tight_layout()
plt.show()
```

# Key Insights & Summary

```python
print("\n ==== KEY INSIGHTS & SUMMARY ==== \n")

# Overall Metrics
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_margin = (total_profit / total_sales) * 100

print("\nOVERALL PERFORMANCE METRICS:\n")
print(f"Total Sales Amount: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%\n")

# Top Performing Category
top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f'\n TOP PERFORMING CATEGORIES:\n')
for i, (category, profit) in enumerate(top_categories.items(), 1):
    print(f"{i}. {category}: ${profit:,.2f}")

#Regional Insights
best_region = df.groupby('Region')['Profit'].sum().idxmax()
print(f"\nBest Performing Region: {best_region}")

# Delivery Performance
delivery_status = df['Delivery Status'].value_counts(normalize=True) * 100
print(f"\nDelivery Status:")
for status,percentage in delivery_status.items():
    print(f"{status}: {percentage:.1f}%")

# Costomer Segment Insights
best_segment = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\nMost profitable Customer Segment:{best_segment}")
```

# Export Result

```python
analysis_results = {
    "total_sales": total_sales,
    "total_profit": total_profit,
    "total_orders": total_orders,
    "avg_profit_margin": avg_profit_margin,
    "top_categories": top_categories,
    "best_region": best_region,
    "best_costomer_segment": best_segment
}

# Convert to DataFrame and Save as CSV
results_df = pd.DataFrame([analysis_results])
results_df.to_csv("EDA_Superstor_Analysis_Results.csv", index=False)
# Exporting the cleaned and analyzed data to a new CSV file
print("\nâœ…Analysis results exported to 'EDA_Superstor_Analysis_Results.csv'")
```
